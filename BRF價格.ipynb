```python
import pandas as pd
csvdata = pd.read_csv("2019_BRF價格.csv",encoding="big5")
csvdata
```

```python
list3 = []
for x in range(0,1223):
    list3.append(csvdata2.iloc[x,2])
list3
```

```python
list2c = []
for i in range(len(list1a)):
    if (list1a[i] == list2b[i] - 2):
        list2c.append(list2[i])
    elif (list1a[i] - 10 == list2b[i]):
        list2c.append(list2[i])
list2c
```

```python
list3a = []
for i in range(len(list1a)):
    if (list1a[i] == list2b[i] - 2):
        list3a.append(float(list3[i]))
    elif (list1a[i] - 10 == list2b[i]):
        list3a.append(float(list3[i]))
list3a       
```

```python
import time,datetime
def get_week_day(date):
    week_day_dict = {
    0 : '星期一',
    1 : '星期二',
    2 : '星期三',
    3 : '星期四',
    4 : '星期五',
    5 : '星期六',
    6 : '星期天',
    }
    day = date.weekday()
    return week_day_dict[day]
#用time,datetime套件來判斷星期幾
```

```python
list4 = []
for i in range(len(list1b)):
    list4.append(get_week_day(datetime.datetime(list1c[i],list1d[i],list1e[i])))
#由於get_week_day(datetime.datetime(2019,5,25))需要以這種填入三個數字的形式，故將年月日分開存成list套入
list4
```

```python
list4a = []
rep1 = ['1' if x == '星期一' else x for x in list4] #以此形式替代list中的元素
rep2 = ['2' if x == '星期二' else x for x in rep1]
rep3 = ['3' if x == '星期三' else x for x in rep2]
rep4 = ['4' if x == '星期四' else x for x in rep3]
rep5 = ['5' if x == '星期五' else x for x in rep4]
for i in range(len(list4)):
    list4a.append(int(rep5[i]))
list4a
```

```python
list5 = []
for i in range(0,282):
    list5.append(list3a[i+1]-list3a[i]) #將第x+1個減掉第x個結算價，即可求出價差
list5.insert(0,0) #會少一個數，故將第一位補0
list5
```

```python
listtotal = zip(list1b,list2c,list3a,list4,list5) #將list鏈起來，ex:a=[1,2,3] b=[z,x,y] zip(a,b)=[(1,z),(2,x),(3,y)]
listtotal
```

```python
csvdata3 = pd.DataFrame(listtotal,columns=["交易日期","到期月份(週別)","結算價","星期","價差"]) 
#前位的list是一個元素去對後面的columns,所以若每[]中有五個元素即可對應五個欄位
for i in range(242,283):
    csvdata3.drop(index = i,inplace = True) #刪除後面錯誤的日期資料
csvdata3
```

```python
Result ='C://Users/k2294/2019_BRF價格_整理版.csv'
csvdata3.to_csv( Result  , index=False , encoding="utf_8_sig" ) #encoding="utf_8_sig"可以防止中文字亂碼
```

```python
list6 = []
calculate = []
total = 0
for i in range(0,242):
    if(list4a[i+1]-list4a[i] > 0):
        calculate.append(list5[i+1]) #如果後一個星期x減掉前一個星期x >0，則將後一個星期x的價差值存入list
    else:
        for j in range(len(calculate)):
            total = total + calculate[j]
        list6.append(format(total/len(calculate) , ".3f")) #如果<0，則將先前list中的所有值加起來並除以其個數，此結果即為週平均
        total = 0
        calculate = [] #結束一次週平均的計算，就要清除變數和list
        calculate.append(list5[i+1]) #若<0，則該價差值需加入下一週的list內計算
list6
```

```python
list7 = []
calculate2 = []
for i in range(0,242):
    if(list4a[i+1]-list4a[i] > 0):
        calculate2.append(list1b[i])
    else:
        calculate2.append(list1b[i])
        list7.append(calculate2[0] + " ~ " + calculate2[-1])
        calculate2 = []
list7
```

```python
list8 = []
calculate3 = []
total2 = 0
for i in range(0,242):
    if(list4a[i+1]-list4a[i] > 0):
        calculate3.append(list3a[i])
    else:
        calculate3.append(list3a[i])
        for j in range(len(calculate3)):
            total2 = total2 + calculate3[j]
        list8.append(format(total2/len(calculate3) , ".3f"))
        total2 = 0
        calculate3 = []
list8
```

```python
listtotal2 = zip(list7,list6,list8)

csvdata4 = pd.DataFrame(listtotal2,columns=["週別","結算價變化量之週平均","結算價週平均"])
csvdata4
```
